```python
import numpy as np
from numpy import asarray
# load and display first image with Matplotlib
from matplotlib import image
from matplotlib import pyplot
```

```python
# load image as pixel array
image1 = image.imread("repics/dog.1.jpg")
image2 = image.imread("repics/dog.2.jpg")
image3 = image.imread("repics/dog.3.jpg")
image4 = image.imread("repics/dog.4.jpg")
image5 = image.imread("repics/dog.5.jpg")
image6 = image.imread("repics/dog.6.jpg")
image7 = image.imread("repics/dog.7.jpg")
image8 = image.imread("repics/dog.8.jpg")
image9 = image.imread("repics/dog.9.jpg")
image10 = image.imread("repics/dog.10.jpg")
```

```python
# convert image to numpy array
image_array1 = asarray(image1)
image_array2 = asarray(image2)
image_array3 = asarray(image3)
image_array4 = asarray(image4)
image_array5 = asarray(image5)
image_array6 = asarray(image6)
image_array7 = asarray(image7)
image_array8 = asarray(image8)
image_array9 = asarray(image9)
image_array10 = asarray(image10)
```

```python
# creating stack using image arrays
SIZE = 10
S = [0] * (SIZE + 1)  # to initialize the list
top = 0
```

```python
S
```

```python
# Function for adding the images
def push(x):
    global top
    top = top + 1
    if top > SIZE:
        print("Stackoverflow")
    else:
        S[top] = x
```

```python
# Function for printing the selected image from stack
def stackprint():
    i = int(input(print("Enter any index value of corresponding stacked image")))
    if i == 1:
        #print("array is", S[1:SIZE + 1])
        print("Displaying elements in stack")
        print("First element is:")
        pyplot.imshow(S[1])
        pyplot.show()
    elif i == 2:
        print("Second element is:")
        pyplot.imshow(S[2])
        pyplot.show()
    elif i == 3:
        print("Third element is:")
        pyplot.imshow(S[3])
        pyplot.show()
    elif i == 4:
        print("Fourth element is:")
        pyplot.imshow(S[4])
        pyplot.show()
    elif i == 5:
        print("Fifth element is:")
        pyplot.imshow(S[5])
        pyplot.show()
    elif i == 6:
        print("Sixth element is:")
        pyplot.imshow(S[6])
        pyplot.show()
    elif i == 7:
        print("Seventh element is:")
        pyplot.imshow(S[7])
        pyplot.show()
    elif i == 8:
        print("Eighth element is:")
        pyplot.imshow(S[8])
        pyplot.show()
    elif i == 9:
        print("Ninth element is:")
        pyplot.imshow(S[1:SIZE+9])
        pyplot.show()
    elif i == 10:
        print("Tenth element is:")
        pyplot.imshow(S[10])
        pyplot.show()

    else:
        print("Invalid Index")
```

```python
# Function for making collage
def display():

    #Slicing each images
    image1_1 = S[1][:199, :100]
    image2_1 = S[2][:199, :100]
    image3_1 = S[3][:199, :100]
    image4_1 = S[4][:199, :100]

    # for collage making
    collage1_1 = np.concatenate((image1_1, image2_1), axis=0)
    print(collage1_1)
    collage1_2 = np.concatenate((image3_1, image4_1), axis=0)
    print(collage1_2)
    collage1 = np.concatenate((collage1_1, collage1_2), axis=0)
    print(collage1)
    pyplot.imshow(collage1)
    pyplot.show()
```

```python
# Driver code
while (True):

    x = int(input(print(
            "Enter your choice \n 1. To add images into stack  \n 2. Enter the index to display the stacked image \n 3. To display the the collage \n 4. Exit")))

    if x == 1:
        push(image_array1)
        push(image_array2)
        push(image_array3)
        push(image_array4)
        push(image_array5)
        push(image_array6)
        print("Stacked the six images!!.....")
    elif x == 2:
        stackprint()
    elif x == 3:
        display()
    elif x == 4:
        break
    else:
        print("Invalid choice")
```
