```
count=0
n1=4
n2=2
ht,wt=100,100
```

```
for j in range(0,n2):
    for i in range(0,n1):
        x1=int((wt/n1)*i)
        y1=int((ht/n2)*j)
        x2 = int(wt / n1)
        y2 = int(ht / n2)
        print(y1,x1,y2,x2)
        img = image[y1:y1+y2,x1:x1+x2] 
        count += 1
        cv2_imshow(img)
        #cv2.waitKey(0)   
```
